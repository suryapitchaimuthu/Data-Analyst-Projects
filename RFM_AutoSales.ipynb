## RFM Analysis

```python
AS_data['ORDERDATE'].max()
```

```python
today_date=dt.datetime(2024,7,2)
```

```python
AS_data.groupby("ORDERNUMBER").agg({"ORDERDATE": lambda OrderDate: (today_date - OrderDate.max()).days, 
                               "QUANTITYORDERED": lambda QuantityOrder: QuantityOrder.nunique(), 
                               "SALES": lambda Sales: Sales.sum()})
```

```python
rfm=AS_data.groupby('ORDERNUMBER').agg({'ORDERDATE' : lambda OrderDate: (today_date - OrderDate.max()).days,
                                   'QUANTITYORDERED': lambda QuantityOrder : QuantityOrder.nunique(),
                                   'SALES' : lambda Sales : Sales.sum()
                                   })
```

```python
rfm.info()
```

```python
rfm.columns=['recency','frequency','monetary']
```

```python
rfm
```

### 4) Calculation of RFM scores

```python
rfm['recency_score']=pd.qcut(rfm['recency'],q=5,labels=[5,4,3,2,1])
rfm['frequency_score']=pd.qcut(rfm['frequency'],q=5, labels=[1,2,3,4,5])
```

```python
rfm['rf_score']=(rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
```

### 5) Segmentation of RF Scores

```python
seg_map={r"[1-2][1-2]" : "hibernating",
         r"[1-2][3-4]" : "At Risk",
         r"[1-2]5" : "Cant Loose",
         r"3[1-2]" : "About to sleep",
         r"33" : "Need attention",
         r"[3-4][4-5]" : "Loyal customers",
         r"41" : "Promising",
         r"51" : "New customers",
         r"[4-5][2-3]" : "Potential Loyalist",
         r"5[4-5]" : "champions"
        }
         
```

```python
rfm['segment']=rfm['rf_score'].replace(seg_map, regex=True)
```

```python
rfm
```

```python
rfm.reset_index(inplace=True)
```

```python
rfm
```

### 6)Analysis and Visualization

```python
rfm_summary=rfm[['segment','recency','frequency','monetary']].groupby('segment').agg(['mean','count'])
```

```python
rfm_summary
```

```python
columns=['recency','frequency','monetary']
for col in columns:
    data=rfm.groupby('segment').agg({col:'mean'})
    plt.bar(data.index,data[col])
    plt.xlabel('Segment')
    plt.ylabel(f'{col}')
    plt.title(f'Average {col} by segment')
    plt.xticks(rotation=90)
    plt.show()
```
